```python
import numpy as np
import matplotlib.pyplot as plt
import mido
from mido import MidiFile
import json
```

```python
midi = MidiFile('samples/midi.mid')
```

```python
fs = 48000
H = 2048
```

```python
melody = []

for msg in midi.play():
    if msg.type == 'note_off':
        time = msg.time
        note = msg.note
        
        for i in range(int(time * fs / H)):
            melody.append(note)
```

```python
t = np.arange(len(melody)) * H / fs
plt.xlabel('Time (s)')
plt.ylabel('Note')
plt.plot(melody)
```

```python
data = {
    'melody': melody
}

with open('midi/ka.json', 'w') as fp:
    json.dump(data, fp, indent=4)
```
